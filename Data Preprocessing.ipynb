### Creating the DataFrame

First, a small-scale test version with only the first couple of wikipeople entries.

```python
data = []
college_temp = ''

college_boolean = False

with open('wikipeople.txt') as file_reader:
    counter = 0
    temp_dict = {}
    for line in file_reader:
        counter += 1
        if (counter > 30000):
            break
        
        r_name = re.search('<title>',line)
        if r_name:
            temp_dict['name'] = line.strip()
        
        r_next = re.search('\|.*=',line)
        if r_next:
            college_boolean = False
        r_education = re.search('\|\s*education\s*=',line)
        if r_education:
            college_boolean = True
        r_alma_mater = re.search('\|\s*alma_mater\s*=',line)
        if r_alma_mater:
            college_boolean = True
        if college_boolean:
            college_temp += line.strip()
        
        r_birth_date = re.search('\|\s*birth_date\s*=',line)
        if r_birth_date:
            temp_dict['birth_date'] = line.strip()
        r_born = re.search('\|\s*born\s*=',line)
        if r_born:
            temp_dict['born'] = line.strip()
        r_occupation = re.search('\|\s*occupation\s*=',line)
        if r_occupation:
            temp_dict['occupation'] = line.strip()
            
        r = re.search('</page>',line)
        if r:
            temp_dict['education'] = college_temp
            data.append(temp_dict)
            temp_dict = {}
            college_temp = ''

wikipeople = pandas.DataFrame(data)
wikipeople = wikipeople.drop([0],axis=0)
wikipeople.head()
```

Now it's time to pull off the gloves and actually create this dataframe.

```python
data = []
college_temp = ''

college_boolean = False

with open('wikipeople.txt') as file_reader:
    temp_dict = {}
    for line in file_reader:
        
        r_name = re.search('<title>',line)
        if r_name:
            temp_dict['name'] = line.strip()
        
        r_next = re.search('\|.*=',line)
        if r_next:
            college_boolean = False
        r_education = re.search('\|\s*education\s*=',line)
        if r_education:
            college_boolean = True
        r_alma_mater = re.search('\|\s*alma_mater\s*=',line)
        if r_alma_mater:
            college_boolean = True
        if college_boolean:
            college_temp += line.strip()
        
        r_birth_date = re.search('\|\s*birth_date\s*=',line)
        if r_birth_date:
            temp_dict['birth_date'] = line.strip()
        r_born = re.search('\|\s*born\s*=',line)
        if r_born and 'birth_date' not in temp_dict.keys():
            temp_dict['birth_date'] = line.strip()
        r_occupation = re.search('\|\s*occupation\s*=',line)
        if r_occupation:
            temp_dict['occupation'] = line.strip()
            
        r = re.search('</page>',line)
        if r:
            temp_dict['education'] = college_temp
            data.append(temp_dict)
            temp_dict = {}
            college_temp = ''

wikipeople = pandas.DataFrame(data)
wikipeople = wikipeople.drop([0],axis=0)
wikipeople.head()
```

### Data Cleaning Part One

```python
wikipeople['name'] = [ re.search('>(.*)<',temp).group(1) for temp in wikipeople['name'] ]
wikipeople['name']
```

What is `Draft:Julie Nagam`? Let's get rid of those.<br>
(Also, we're going to start working with wikipeople_filtered to leave the earlier dataset intact.)

```python
df1 = wikipeople[wikipeople['name'].str.match('Draft:')]
wikipeople_filtered = wikipeople.drop(df1.index)
wikipeople_filtered
```

```python
def nanToEmpty(thing):
    if thing is numpy.nan:
        return ''
    else:
        return thing
```

```python
wikipeople_filtered['birth_date'] =  [ nanToEmpty(temp) for temp in wikipeople_filtered['birth_date'] ]
wikipeople_filtered
```

```python
re.search('([0-9]{4})',wikipeople_filtered['birth_date'][1]).group(1)
```

```python
wikipeople_filtered['birth_date'] = [ re.sub('.*([0-9]{4}).*','\\1',temp) for temp in wikipeople_filtered['birth_date'] ]
```

Next, let's make the `education` column array lists of college names rather than ugly strings.

```python
education_filtered = []
for person in wikipeople_filtered['education']:
    temp_array = re.findall('\[\[([^\[]*)\]\]',person)
    education_filtered.append(temp_array)
education_filtered[:5]
```

```python
wikipeople_filtered['education'] = education_filtered
wikipeople_filtered.head()
```
